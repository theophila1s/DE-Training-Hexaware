```python
#1.Loading Data in Pandas DataFrame
import pandas as pd

df = pd.read_csv("C:/Users/theop/OneDrive/Desktop/DE TRAINING/Python/LoanData.csv")
df.head()  # Displays first 5 rows
```

```python
# 5. Descriptive Statistical Measures of the DataFrame

print(df.describe())               # Summary for numeric columns
print(df.describe(include='all')) 
```

```python
#6. Missing Data Handling

# Count of missing values in each column
print(df.isnull().sum())

# Fill missing values with 0
df_filled = df.fillna(0)

# Drop rows with missing values
df_dropped = df.dropna()

# Optionally: fill with column mean
# df['column_name'] = df['column_name'].fillna(df['column_name'].mean())
```

```python
#7. Sorting DataFrame Values

# Example: Sorting by 'LoanAmount' column (adjust based on actual column name)
df_sorted = df.sort_values(by='LoanAmount', ascending=False)
print(df_sorted[['Loan_ID', 'LoanAmount']].head())
```

```python
#8. Merging DataFrames

# Dummy example for merge
df1 = pd.DataFrame({'Loan_ID': ['LP001002', 'LP001003'], 'Extra_Info': ['A', 'B']})
merged_df = pd.merge(df, df1, on='Loan_ID', how='left')
print(merged_df.head())
```

```python
#9. Apply Function

# Example: Categorizing income levels (adjust column if needed)
def income_group(income):
    if income < 2500:
        return 'Low'
    elif income < 6000:
        return 'Medium'
    else:
        return 'High'

df['IncomeGroup'] = df['ApplicantIncome'].apply(income_group)
print(df[['ApplicantIncome', 'IncomeGroup']].head())
```

```python
#10. By Using the Lambda Operator

# Multiply LoanAmount by 1000 using lambda
df['LoanAmount_Updated'] = df['LoanAmount'].apply(lambda x: x * 1000 if pd.notnull(x) else x)
print(df[['LoanAmount', 'LoanAmount_Updated']].head())
```

```python
#11. Visualizing DataFrame

import seaborn as sns
import matplotlib.pyplot as plt

# Countplot of Loan_Status
sns.countplot(data=df, x='Loan_Status')
plt.title("Loan Approval Count")
plt.show()

# Distribution of Loan Amount
plt.figure(figsize=(8,4))
sns.histplot(df['LoanAmount'].dropna(), bins=30, kde=True)
plt.title("Loan Amount Distribution")
plt.show()
```
